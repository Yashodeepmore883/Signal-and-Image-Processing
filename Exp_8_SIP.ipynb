```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
structuring_element_sizes = [11,15,45]
```

```python
image = cv2.imread('/content/OIP.jpg')
if image is None:
  print("Error: Unable to load image.")
  exit()
structuring_element_sizes = [11,15,45]
```

```python
eroded_images = []
for size in structuring_element_sizes:
  kernel = np.ones((size,size),np.uint8)
  eroded = cv2.erode(image, kernel, iterations=1)
  eroded_images.append(eroded)
```

```python
plt.figure(figsize=(15,5))
for i in range(len(structuring_element_sizes)):
  plt.subplot(1, len(structuring_element_sizes), i+1)
  plt.imshow(eroded_images[i], cmap='gray')
  plt.title(f'Size:{structuring_element_sizes[i]}x{structuring_element_sizes[i]}')
  plt.axis('off')
  plt.show()
```

```python
image = cv2.imread('/content/OIP.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5,5),np.uint8)
```

```python
dilated_image = cv2.dilate(image, kernel, iterations=1)
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(dilated_image, cmap='gray')
plt.title('Dilated Image')
plt.axis('off')

plt.show()
```
