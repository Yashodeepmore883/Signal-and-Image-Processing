```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
image = cv2.imread('/content/image-10.png', cv2.IMREAD_GRAYSCALE)
```

```python
median_filtered = cv2.medianBlur(image, 5)
```

```python
_, binarized = cv2.threshold(median_filtered,0,255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
kernel_opening = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(binarized, cv2.MORPH_OPEN, kernel_opening, iterations=2)
```

```python
kernel_closing = np.ones((5,5), np.uint8)
closing = cv2.morphologyEx(binarized, cv2.MORPH_CLOSE, kernel_closing, iterations=3)
```

```python
cv2_imshow(image)
cv2_imshow(closing)
```
