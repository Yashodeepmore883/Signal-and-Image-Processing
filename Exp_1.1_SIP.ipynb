```python
from matplotlib import style
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='line one',linewidth=2)
plt.plot(x2,y2,'b',label='line two',linewidth=2)
plt.title('Epic Info')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Trend one", color='g')
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Trend two", color='b')

plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Comparative plot')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
circles = plt.imread("data/circles.jpg")
plt.imshow(circles)
plt.show()
```

```python
patch = (300,200)
center =(190,220)
cropped_image = cv2.getRectSubPix(circles,patch,center)
plt.imshow(cropped_image)
plt.show()
```

```python
def crop(img,center,width,height):
  return cv2.getRectSubPix(img, (width,height),center)
```

```python
cropped = crop(circles, center=(190,220), width=300, height=200)
plt.imshow(cropped)
plt.show()
```
