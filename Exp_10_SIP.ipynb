```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
image = cv2.imread('/content/OIP.jpg')
if image is None:
    print("Error: Unable to load image.")
    exit()
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F,1,0,ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F,0,1,ksize=3)
```

```python
abs_sobel_x = np.abs(sobel_x)
abs_sobel_y = np.abs(sobel_y)
```

```python
combined_gradients = cv2.addWeighted(abs_sobel_x,0.5,abs_sobel_y,0.5,0)
```

```python
cv2_imshow(abs_sobel_x.astype(np.uint8))
cv2_imshow(abs_sobel_y.astype(np.uint8))
cv2_imshow(combined_gradients.astype(np.uint8))
```
