```python
from PIL import Image
original_image_path= r'/content/believe.jpg'
original_image=Image.open(original_image_path)
original_image.show
display(original_image)
```

```python
import matplotlib.pyplot as plt
```

```python
def apply_grayscale_resolution(original_image_path,output_image_path,bits_per_pixel):
 original_image_path=Image.open(original_image_path)
 grayscale_image=original_image.convert("L",palette=Image.ADAPTIVE,colors=2**bits_per_pixel)
 grayscale_image.save(output_image_path)
 compression_ratio = 8/bits_per_pixel
 fig, axs = plt.subplots(1,2,figsize=(10,5))

 axs[0].imshow(original_image)
 axs[0].set_title("Original Image")
 axs[0].axis("off")

 axs[1].imshow(grayscale_image,cmap="gray")
 axs[1].set_title(f"Grayscale Image (Bits per Pixel: {bits_per_pixel})")
 axs[1].axis("off")

 plt.show()

 return compression_ratio

original_image_path = r'/content/believe.jpg'
output_image_path = r'/content/grayscale_image.jpg'
bits_per_pixel = 2

compression_ratio = apply_grayscale_resolution(original_image_path,output_image_path,bits_per_pixel)
print(f"compression ratio: {compression_ratio}")
```
