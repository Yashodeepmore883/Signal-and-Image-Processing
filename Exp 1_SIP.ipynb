```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
A = 3
freq = 5
```

```python
t = np.linspace(0,1,1000)
```

```python
t
```

```python
x = A*np.cos(2*np.pi*freq*t)
```

```python
plt.figure(figsize=(10,10))
```

```python
plt.plot(t,x,'-r')
plt.xlabel('Time in seconds',fontsize=15)
plt.ylabel('Amplitude',fontsize=15)
plt.title('Cosine signal with frequency 5 Hz',fontsize=20)
plt.grid()
```

```python
t = np.linspace(0,1,1000)
```
