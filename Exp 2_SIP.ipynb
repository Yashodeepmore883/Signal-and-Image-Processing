```python
t=np.arange(0,10,1)
a=np.exp(t)
b=np.exp(-t)
plt.subplot(2,2,1)
plt.plot(t,b)
plt.subplot(2,2,2)
plt.stem(t,b)
plt.subplot(2,2,3)
plt.plot(t,a)
plt.subplot(2,2,4)
plt.stem(t,a)
```

```python
n = np.arange(0,6)
x = np.array([1,2,3,4,5,6])
plt.subplot(1,2,1)
plt.stem(n, x)
plt.axis([-1,6,0,7])
l = []
for i in range(0, 6):
    if i % 2 == 0:
        l.append(x[i])
do = np.array(l)
n_downsampled = np.arange(0,len(do) * 2,2)
plt.subplot(1, 2, 2)
plt.stem(n_downsampled, do)
plt.axis([-1, 6, 0, 7])
```
